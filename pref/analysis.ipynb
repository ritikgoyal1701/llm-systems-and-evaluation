```python
import pandas as pd

df = pd.read_csv("perf/metrics.csv")

print("P50 latency:", df["latency"].quantile(0.5))
print("P95 latency:", df["latency"].quantile(0.95))
print("P99 latency:", df["latency"].quantile(0.99))

print("Avg TTFT:", df["ttft"].mean())
print("Avg TPOT:", df["tpot"].mean())
```
